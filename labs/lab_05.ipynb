1-. Lea el "total_profit" de todos los meses, muéstrelo usando un gráfico lineal y un gráfico de dispersión.

**Se observa la tabla de month_number y total_profit

```python
df.loc[:,['month_number','total_profit']]
```

```python
import numpy as np
```

```python
# gráfico simple

# datos
x = df.month_number
y = df.total_profit
# gráfico

# tamaño del gráfico
fig = plt.figure(figsize=(10, 5)) 

# graficar
plt.plot(
    x, # eje x
    y, # eje y
    label='linea', # etiquetado
    color="red", # color
    linewidth=1 # tamaño de la curva
)

plt.xticks(np.arange(0, 13, 1.0))#se usa para que se vea mejor el eje x

plt.legend() # agregar etiquetado
plt.title("grafico simple") # agregar titulo
plt.xlabel('month_number') # nombre eje x
plt.ylabel('total_profit') # nombre eje y
plt.grid() # agregar grillado
plt.show() # mostrar gráfico
```

```python
# gráfico de dispersión

# datos
x = df.month_number
y = df.total_profit
s = 300

# gráfico
fig = plt.figure(figsize=(10, 5)) # ventana
plt.scatter(
    x, # eje x
    y, # eje y
    s, # tamaño de los puntos
    c="r", # color
    alpha=0.7, # intensidad color
    marker='.', # forma de los puntos
    label="puntos" # etiquetado fijando posición
)
plt.xticks(np.arange(0, 13, 1.0))#se usa para que se vea mejor el eje x
plt.title("gráfico de dispersión") # agregar titulo
plt.xlabel('month_number') # nombre eje x
plt.ylabel('total_profit') # nombre eje y
plt.legend(loc='upper left')
plt.show()# mostrar gráfico
```

2-. Lea todos los datos de ventas de productos y muéstrelos utilizando un gráfico multilínea.

**Se observa la tabla con los valores a graficar**

```python
df.loc[:,['month_number','facecream','facewash','toothpaste','bathingsoap','shampoo','moisturizer']]
```

**Solo se observan 5 lineas pues facewash y moisturizer tienen los mismos valores**

```python
# gráfico compuesto

# datos
x = df.month_number

# gráfico
# tamaño del gráfico
fig = plt.figure(figsize=(10, 5)) 

# graficar

# facecream
y=df.facecream
plt.plot(
    x, # eje x
    y, # eje y
    label='facecream', # etiquetado
    color="g", # color
    linewidth=1 # tamaño de la curva
)
# facewash
y=df.facewash
plt.plot(
    x, # eje x
    y, # eje y
    label='facewash', # etiquetado
    color="b", # color
    linewidth=1 # tamaño de la curva
)
# toothpaste
y=y=df.toothpaste
plt.plot(
    x, # eje x
    y, # eje y
    label='toothpaste', # etiquetado
    color="r", # color
    linewidth=1 # tamaño de la curva
)
# bathingsoap
y=df.bathingsoap
plt.plot(
    x, # eje x
    y, # eje y
    label='bathingsoap', # etiquetado
    color="black", # color
    linewidth=1 # tamaño de la curva
)
# shampoo
y=df.shampoo
plt.plot(
    x, # eje x
    y, # eje y
    label='shampoo', # etiquetado
    color="y", # color
    linewidth=1 # tamaño de la curva
)
# moisturizer 
y=df.moisturizer
plt.plot(
    x, # eje x
    y, # eje y
    label='moisturizer', # etiquetado
    color="purple", # color
    linewidth=1 # tamaño de la curva
)
plt.xticks(np.arange(0, 13, 1.0))#se usa para que se vea mejor el eje x
plt.legend() # agregar etiquetado
plt.title("grafico compuesto") # agregar titulo
plt.xlabel('month_number') # nombre eje x
plt.ylabel('cantidad vendida') # nombre eje y
plt.grid() # agregar grillado
plt.show() # mostrar gráfico
```

3-. Lea los datos de ventas de productos de "facecream" y "facewash" y muéstrelos usando el gráfico de barras.

**Se observa la tabla con los valores de los cuales se realizara el gráfico de barras**

```python
df.loc[:,['month_number','facecream','facewash']]
```

**Entendi que habia que hacer ambos gráficos de barra en el mismo por lo que se diferencian con colores**

```python
# datos
x = df.month_number

# gráfico
fig = plt.figure(figsize=(10, 5))

y = df.facecream
plt.bar(
    x, # eje x
    y, # eje y 
    align='center', # centrar nombre eje y
    color="r", # color 
    alpha=1 # intensidad del color
)

y = df.facewash
plt.bar(
    x, # eje x
    y, # eje y 
    align='center', # centrar nombre eje y
    color="b", # color 
    alpha=1 # intensidad del color
)
plt.xticks(np.arange(0, 13, 1.0))#se usa para que se vea mejor el eje x
plt.xlabel('month_number') # nombre eje x
plt.ylabel('numero ventas') # nombre eje y
plt.show()# mostrar gráfico
```

**Este primer gráfico es sin incluir total_units y total_profit pues hacen que sea dificil observar el gráfico**

```python
# Pre-format DataFrame
s_df = df.drop(['month_number','total_units','total_profit'], axis=1)
# New boxplot using s_df
plt.figure(figsize=(10, 6))
sns.boxplot(data=s_df)
plt.show()# mostrar gráfico
```

```python
# Pre-format DataFrame
s_df = df.drop(['month_number'], axis=1)
# New boxplot usando s_df
plt.figure(figsize=(10, 6))
sns.boxplot(data=s_df)
plt.show()# mostrar gráfico
```

5-. Calcule los datos de ventas totales del año pasado para cada producto y muéstrelos usando un gráfico circular

```python
#gráfico circular a.k.a. gráfico de pastel

#Se suman las ventas de cada producto
a=sum(df.facecream)
b=sum(df.facewash)
c=sum(df.toothpaste)
d=sum(df.bathingsoap)
e=sum(df.shampoo)
f=sum(df.moisturizer)
#se generan los arreglos con los cuales se creara el gráfico
ventas_totales = [a,b,c,d,e,f]
productos = ["facecream","facewash","toothpaste","bathingsoap","shampoo","moisturizer"]
#se crea el gráfico
plt.pie(ventas_totales, labels=productos, autopct='%1.2f%%')#se crea el gráfico, el cual incluye el porcentaje de las centas de cada producto
plt.show()# mostrar gráfico
```
